```python
fraud = df['Class'].value_counts()[1]
normal = df['Class'].value_counts()[0]

fraud_percentage = (fraud/(fraud+normal))*100

fraud_percentage
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Class", y="Amount", data=df)

plt.title("Transaction Amount vs Fraud")

plt.show()
```

```python
plt.figure(figsize=(15,10))

corr = df.corr()

sns.heatmap(corr, cmap="coolwarm")

plt.title("Correlation Matrix")

plt.show()
```

```python
corr_with_target = df.corr()['Class'].sort_values(ascending=False)

corr_with_target
```

```python
corr_with_target.head(10).plot(kind='bar')

plt.title("Top Correlated Features with Fraud")

plt.show()
```
